```python
#from glove import Corpus, Glove
import nltk
from nltk.corpus import brown
nltk.download('brown')
from collections import Counter
import numpy as np
from scipy.sparse import csr_matrix
import pickle
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

brown_sentences = list(brown.sents())

test_sentences = [
    "the teacher is teaching students about grammar".split(),
    "she teaches mathematics at the university".split(),
    "the teacher prepared teaching materials yesterday".split(),
    "many teachers attend teaching conferences annually".split(),
    "effective teaching requires good communication skills".split(),
    "the unteachable student refused to learn".split(),
    "computational linguistics combines computer science and language".split(),
    "the computation took several hours to complete".split(),
    "we computed the results using advanced algorithms".split(),
    "modern computers can compute complex calculations quickly".split(),
    "the recomputation was necessary after finding errors".split(),
    "natural language processing is fascinating".split(),
    "the nature of language is complex".split(),
    "naturally occurring patterns in text are important".split(),
    "the unnaturalness of the translation was obvious".split()
]

all_sentences = brown_sentences + test_sentences
print(f"all sentences: {len(all_sentences)}")
```

```python
# co-occurrence matrix

def build_matrix(sentences, window=5, min_count=1):
    vocab = {}
    word_counts = Counter()
    cooc = Counter()

    print("Computing words frequencies...")
    for sent in sentences:
        for w in sent:
            word_counts[w] += 1

    idx = 0
    for w, cnt in word_counts.items():
        if cnt >= min_count:
            vocab[w] = idx
            idx += 1

    print(f"Vocab size: {len(vocab)}")

    print("Building coocurrence matrix...")
    for sent in tqdm(sentences, desc="sentences"):
        for i, center_word in enumerate(sent):
            if center_word not in vocab:
                continue
            center_id = vocab[center_word]
            left = sent[max(0, i-window):i]
            right = sent[i+1:i+window+1]
            context = left + right

            for j, ctx_word in enumerate(context):
                if ctx_word not in vocab:
                    continue
                ctx_id = vocab[ctx_word]
                dist = window - abs(j - (len(left) if j >= len(left) else j))
                weight = 1.0 / dist
                cooc[(center_id, ctx_id)] += weight
                cooc[(ctx_id, center_id)] += weight  # symmetric

    row, col, data = [], [], []
    for (i,j), val in cooc.items():
        row.append(i)
        col.append(j)
        data.append(val)

    cooc_matrix = csr_matrix((data, (row, col)), shape=(len(vocab), len(vocab)))
    return cooc_matrix, vocab

cooc_matrix, vocab = build_matrix(all_sentences, window=5, min_count=1)
print(f"Matrix shape: {cooc_matrix.shape}, non-zeros: {cooc_matrix.nnz}")
```

```python
# Train

from glove import Glove

glove = Glove(no_components=300, learning_rate=0.05)
print("GloVe tarining(20epochs)...")
glove.fit(cooc_matrix, epochs=20, verbose=True)

# add vocabulary
id_to_word = {i: w for w, i in vocab.items()}
glove.dictionary = vocab
glove.word_to_id = vocab
glove.id_to_word = id_to_word

# most similar func
def glove_most_similar(word, topn=5):
    if word not in vocab:
        return []
    idx = vocab[word]
    vec = glove.word_vectors[idx]
    sims = np.dot(glove.word_vectors, vec) / (np.linalg.norm(glove.word_vectors, axis=1) * np.linalg.norm(vec) + 1e-8)
    top_idx = np.argsort(-sims)[1:topn+1]
    return [(id_to_word[i], float(sims[i])) for i in top_idx]

with open('glove_brown_custom.pkl', 'wb') as f:
    pickle.dump(glove, f)

print("GloVe model saved!")
```

```python
# 4 Word2Vec and FastText

from gensim.models import Word2Vec, FastText

word2vec = Word2Vec.load("word2vec_brown.model")
fasttext = FastText.load("fasttext_brown.model")
```

Experiments

```python
# Բոլոր experiment-ները միասին

oov_words = ['teachable', 'unteacher', 'supercomputer', 'miscomputation', 'unnaturally']
rare_pairs = [("unteachable", "teacher"), ("recomputation", "computation"), ("unnaturalness", "natural")]

print("=== OOV ===")
for w in oov_words:
    print(f"{w:<18} W2V: {'✓' if w in word2vec.wv else '✗'}   "
          f"FT: {'✓' if w in fasttext.wv else '✗'}   "
          f"GloVe: {'✓' if w in glove.dictionary else '✗'}")

print("\n=== Rare words similarity ===")
for rare, common in rare_pairs:
    w2v = word2vec.wv.similarity(rare, common) if rare in word2vec.wv else 0
    ft = fasttext.wv.similarity(rare, common)
    gl = glove_most_similar(rare)[0][1] if rare in glove.dictionary else 0
    print(f"{rare} → {common}:  W2V {w2v:.4f} | FT {ft:.4f} | GloVe {gl:.4f}")

print("\n=== teaching top-5 ===")
print("Word2Vec:", [x[0] for x in word2vec.wv.most_similar('teaching', topn=5)])
print("FastText:", [x[0] for x in fasttext.wv.most_similar('teaching', topn=5)])
print("GloVe:   ", [x[0] for x in glove_most_similar('teaching', 5)])

print("\n=== computation top-5 ===")
print("Word2Vec:", [x[0] for x in word2vec.wv.most_similar('computation', topn=5)])
print("FastText:", [x[0] for x in fasttext.wv.most_similar('computation', topn=5)])
print("GloVe:   ", [x[0] for x in glove_most_similar('computation', 5)])
```
